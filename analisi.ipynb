```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.ticker import ScalarFormatter
```

```python
# Variabili globali

path_old = ["experiments/obdf-resultAnalysis/resultFiles_old/20000.csv","experiments/obdf-resultAnalysis/resultFiles_old/200000.csv"]

path_new = ["experiments/obdf-resultAnalysis/resultFiles-Last/20000.csv","experiments/obdf-resultAnalysis/resultFiles-Last/200000.csv"]
```

```python
all_result = []
all_result_old = []

for a,b in zip(path_new,path_old):
    all_result.append(pd.read_csv(a))
    all_result_old.append(pd.read_csv(b))
```

```python
# Rimozione della colonna in pi√π per tutti i dataframe

for a in range(0,len(path_old)):
    all_result[a] = all_result[a].loc[:,~all_result[a].columns.str.match("Unnamed")]
    all_result_old[a] = all_result_old[a].loc[:,~all_result_old[a].columns.str.match("Unnamed")]
```

```python
# Andiamo a stampare il barplot con i risultati

ax = all_result[0].plot.bar(rot=0,figsize=(12,8))
ax.set_xticks(range(0,13),all_result[0]['query'].values)
ax.set_yscale('log')
ax.yaxis.set_major_formatter(ScalarFormatter())
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=8)
```

```python
for num,acqui in zip(range(0,len(path_old)),['20000','200000']):

    ax = all_result[num].plot.bar(rot=0,figsize=(12,8))
    ax.set_xticks(range(0,13),all_result[0]['query'].values)
    ax.set_title('Analisi del nuovo datset da '+acqui)
    ax.set_xlabel('Query')
    ax.set_ylabel('tempo medio di esecuzione su scala log.')
    ax.set_yscale('log')
    ax.xaxis.set_label_coords(.08, -.05)
    ax.yaxis.set_major_formatter(ScalarFormatter())
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),
            fancybox=True, shadow=True, ncol=8,title='profili')
```

```python
for num,acqui in zip(range(0,len(path_new)),['20000','200000']):

    ax = all_result_old[num].plot.bar(rot=0,figsize=(12,8))
    ax.set_xticks(range(0,13),all_result[0]['query'].values)
    ax.set_title('Analisi del vecchio datset da '+acqui)
    ax.set_xlabel('Query')
    ax.set_ylabel('tempo medio di esecuzione su scala log.')
    ax.set_yscale('log')
    ax.xaxis.set_label_coords(.08, -.05)
    ax.yaxis.set_major_formatter(ScalarFormatter())
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.08),
            fancybox=True, shadow=True, ncol=8,title='profili')
```
